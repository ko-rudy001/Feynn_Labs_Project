```python
import re

# Remove non-numeric characters from the 'Price' column
ev_up['Ex-Showroom_Price'] = ev_up['Ex-Showroom_Price'].apply(lambda x: ''.join(filter(str.isdigit, str(x))))

# Convert the 'Price' column to float
ev_up['Ex-Showroom_Price'] = ev_up['Ex-Showroom_Price'].astype(float)

ev_up
```

```python
ev_up[ev_up['Make'] == 'Mg']
```

```python
ev_up.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a new column which groups cars based on their price
def price_segments(x):
    if x <= 800000:
        return 'E'
    elif 800000 < x <= 1500000:
        return 'D'
    elif 1500000 < x <= 2500000:
        return 'C'
    elif 2500000 < x <= 3500000:
        return 'B'
    else:
        return 'A'

ev_up['Price_Group'] = ev_up['Ex-Showroom_Price'].apply(price_segments)

# Plot count of each price group
plt.figure(figsize=(12, 6))
sns.countplot(x=ev_up['Price_Group'])
plt.title('Count of Each Price Group')
plt.show()
```

```python
a = len(ev_up[ev_up['Price_Group'] == 'A'].value_counts())
c = len(ev_up[ev_up['Price_Group'] == 'C'].value_counts())
d = len(ev_up[ev_up['Price_Group'] == 'D'].value_counts())
```

```python
len(ev_up['Ex-Showroom_Price'].value_counts()) == (a+c+d)
```

So, majority of vehicles are buyed from group A&d.
A --> vehicles with price greater than Rs. 35,00,000.
D --> vehicles with price greater than Rs. 8,00,000 and smaller than Rs. 15,00,000

```python
df = pd.read_csv('EV Count per State.csv')
```

```python
df.head()
```

```python
df.shape
```

top five states with the highest number of electric vehicles?

```python
# Select the relevant columns
df_selected = df[['State/UT-wise', 'Number of Electric/Hybrid Vehicle (approx.)']]
```

```python
# Sort the data by the number of electric/hybrid vehicles in descending order
df_sorted = df_selected.sort_values(by='Number of Electric/Hybrid Vehicle (approx.)', ascending=False)

# dropping the last row i.e, Total
df_sorted.drop(df.index[-1], inplace=True)

df_sorted
```

```python
# Select the top 5 states/UTs with the highest number of electric/hybrid vehicles
top_5_states = df_sorted.head(5)

# Create a bar plot (vertically)
plt.figure(figsize=(12, 6))
sns.barplot(x='State/UT-wise', y='Number of Electric/Hybrid Vehicle (approx.)', data=top_5_states)
plt.title('Top 5 States/UTs with Highest Number of Electric/Hybrid Vehicles')
plt.xlabel('State/UT')
plt.ylabel('Number of Electric/Hybrid Vehicles')
plt.show()
```

```python
# Select the top 5 states/UTs with the highest number of electric/hybrid vehicles
bottom_5_states = df_sorted.tail(5)

# Create a bar plot (vertically)
plt.figure(figsize=(12, 6))
sns.barplot(x='State/UT-wise', y='Number of Electric/Hybrid Vehicle (approx.)', data=bottom_5_states)
plt.title('Bottom 5 States/UTs with Lowest Number of Electric/Hybrid Vehicles')
plt.xlabel('State/UT')
plt.ylabel('Number of Electric/Hybrid Vehicles')
plt.show()
```
